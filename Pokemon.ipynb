```python
import tensorflow as tf
from keras_preprocessing.image import ImageDataGenerator
```

```python
train_transform = ImageDataGenerator(
    rescale = 1/255,
    shear_range=0.3,
    rotation_range=30,
    zoom_range=0.3,
    width_shift_range=0.3,
    height_shift_range=0.3,
    horizontal_flip=True,    
)
```

```python
train_set = train_transform.flow_from_directory(
    r'D:\ML DATASETS CNN AND ANN\pika_pika\training_set',
    batch_size=32,
    target_size=(64,64),
    class_mode='binary'    
)
```

```python
test_transform = ImageDataGenerator(
    rescale=1/255
)
```

```python
test_set = test_transform.flow_from_directory(
    r'D:\ML DATASETS CNN AND ANN\pika_pika\test_set',
    batch_size=32,
    target_size=(64,64),
    class_mode='binary'
)
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3,activation='relu', input_shape=(64,64,3)))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,activation='relu',kernel_size=3))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dropout(0.2))
```

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(train_set,validation_data = test_set, epochs=25)
```

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img(r'D:\ML DATASETS CNN AND ANN\pika_pika\single_prediction\check_2.png', target_size=(64,64))
test_image = np.expand_dims(test_image,axis=0)
result = cnn.predict(test_image)
# print(train_set.class_indices)
if (result[0][0] == 0):
    print('Raichu')
else:
    print('Pikachu')
```
